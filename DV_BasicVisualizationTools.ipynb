```python
# create a stacked area plot of the 5 countries that contributed the 
# least to immigration to Canada from 1980 to 2013. 
# Use a transparency value of 0.45.

# bottom5 = df_can.sort_values(['Total'],axis=0).head()
# or use below
bottom5 = df_can.nsmallest(5,'Total')

#bottom5 = bottom5.loc[bottom5.index,years]
#bottom5= bottom5.transpose()
# or use below
bottom5 = bottom5[years].transpose()

bottom5.plot(kind='area',alpha=0.45,figsize=(20,10))
plt.title('Trend between bottom 5 contributors to immigration')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.show()
```

### Histogram

```python
df_can['2013'].head()
```

```python
df_can['2013'].plot(kind='hist')

plt.title('Histogram of Immigration from 195 countries in 2013')
# add y-label
plt.ylabel('Number of Countries')
# add x-label
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
count, bin_edges = np.histogram(df_can['2013']) # by default it breaks the dataset into 10 bins
print(count)
print(bin_edges)
df_can['2013'].plot(kind='hist',xticks = bin_edges,figsize =(10,7))

plt.title('Histogram of Immigration from 195 countries in 2013')
# add y-label
plt.ylabel('Number of Countries')
# add x-label
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
# immigration distribution for Denmark, Norway, and Sweden for years 1980 - 2013

df2 = df_can.loc[['Denmark', 'Norway', 'Sweden'],years].transpose()
df2.plot(kind='hist',figsize=(10,10))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
count, binss = np.histogram(df2,15)

df2.plot(kind='hist',
        figsize=(10,7),
        bins=15,
        alpha=0.6,
        xticks=binss,
        # stacked=True,
        color=['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
#immigration distribution for Greece, Albania, and Bulgaria for years 1980 - 2013? 
# Use an overlapping plot with 15 bins and a transparency value of 0.35

df2 = df_can.loc[['Greece', 'Albania', 'Bulgaria'],years].transpose()

count, binss = np.histogram(df2,15)

df2.plot(kind='hist',
        figsize=(10,7),
        bins=15,
        alpha=0.35,
        xticks=binss,
        color=['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Histogram of Immigration from Greece, Albania, and Bulgaria  from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

### Bar charts

```python
dfi = df_can.loc['Iceland', years]

dfi.plot(kind='bar',figsize=(10,6),rot=90)

plt.title('Icelandic immigrants to canada from 1980 to 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

# Annotate arrow
plt.annotate('',  # s: str. Will leave it blank for no text
             xy=(32, 70),  # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),  # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',  # will use the coordinate system of the object being annotated
             arrowprops=dict(arrowstyle='simple', connectionstyle='arc3', color='blue', lw=2)
             )

# Annotate Text
plt.annotate('2008 - 2011 Financial Crisis',  # text to display
             xy=(28, 30),  # start the text at at point (year 2008 , pop 30)
             rotation=73,  # based on trial and error to match the arrow
             va='bottom',  # want the text to be vertically 'bottom' aligned
             ha='left',  # want the text to be horizontally 'left' algned.
             )

plt.show()
```

```python
#top15 = df_can.sort_values(by='Total', ascending=False)
#top15 = top15['Total'].head(15)
# or use below
top15 = df_can.nlargest(15,'Total')['Total']

top15.plot(kind='barh',figsize=(12,12))

plt.xlabel('Number of Immigrants')
plt.title('Top 15 Conuntries Contributing to the Immigration to Canada between 1980 - 2013')

# annotate value labels to each country
for index, value in enumerate(top15): 
    label = format(int(value), ',') # format int with commas
    
    # place text at the end of bar (subtracting 47000 from x, and 0.1 from y to make it fit within the bar)
    plt.annotate(label, xy=(value - 50000, index - 0.1), color='white')

plt.show()
```
